```python
# undersampling
fraud.shape
```

```python
legal_sample = legal.sample(n=473)
```

```python
legal_sample.shape
```

```python
data_undersample = pd.concat([legal_sample,fraud],ignore_index=True)
```

```python
data_undersample
```

```python
data_undersample.shape
```

```python
data_undersample.Class.value_counts()
```

```python
x = data_undersample.drop(columns=['Class'])
y = data_undersample['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler = StandardScaler()
```

```python
x_train_scale= scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scale,y_train)
```

```python
pred = model.predict(x_test_scale)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred , y_test)
```
